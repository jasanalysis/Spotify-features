```python
most_popular = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
```

```python
most_popular
```

```python
df_tracks.set_index('release_date', inplace = True)

df_tracks.index = pd.to_datetime(df_tracks.index)

df_tracks.head()
```

```python
df_tracks[['artists']].iloc[18]
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x : round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis = 1)
```

```python
df_tracks.head()
```

```python
corr_df = df_tracks.drop(['key','mode','explicit'], axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df,annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0)
heatmap.set_title('Correlation between two variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
sample_df
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set_title('Loudness vs Energy')
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'c').set_title('Popularity vs Acousticness')
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
# we will create a distribution plot to visualize total sounds each year

sns.displot(years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'No of songs per year')
```

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax).set(title = 'Years vs duration')
plt.xticks(rotation = 90)
```

```python
df_genre = pd.read_csv(r'C:\Users\jaski\Downloads\spot.csv')
df_genre.head()
```

```python
plt.title('Duration of the song in different genres')
sns.color_palette('pastel',as_cmap = False )
sns.barplot(y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel('Duration in milliseconds')
plt.ylabel('genre')
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize=(10,6))
famous = df_genre.sort_values('popularity', ascending = False).head(10)

sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = 'Most loved genres')
```
